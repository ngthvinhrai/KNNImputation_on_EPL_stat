```python
def nan_euclidean(x1, x2):
  weight = 1  # weight = no. of coordinates / no. of available coordinates
  dist = np.sqrt(np.sum((x1-x2)**2, axis=1))
  return dist
```

```python
def findmax(arr):
  max = arr[0]
  idx = 0
  for i in range(len(arr)):
    if arr[i]>max:
      max = arr[i]
      idx = i
  return max, idx
```

```python
def kmin(arr, k):
  temp = arr[0:k]
  pos = [i for i in range(0,k)]
  max, idx = findmax(temp)
  for i in range(k, len(arr)):
    if arr[i]<max:
      temp[idx] = arr[i]
      pos[idx] = i
      max, idx = findmax(temp)

  return temp, pos
```

```python
def KNNImputation(dataframe, k_neighbours):
  avPoint = dataframe.dropna(axis='rows').drop(['HDP', 'ADP'], axis=1).to_numpy()
  naPoint = dataframe[dataframe['HDP'].isna()].drop(['HDP', 'ADP'], axis=1).to_numpy()
  result = []
  for i in range(len(naPoint)):
    dist = nan_euclidean(naPoint[i], avPoint)
    topdist, pos = kmin(dist, k_neighbours)
    result.append([naPoint[i], topdist, pos])

  return avPoint, naPoint, result
```
